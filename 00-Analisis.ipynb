### Missing data

```python
def get_missing_columns(df):
    return list(df.columns[df.describe(include = 'all').loc['count']<len(df)])
```

```python
df_tables = ['df_genre','df_info','df_item','df_occupation','df_test','df_train','df_user']
```

```python
display('df_genre',get_missing_columns(df_genre))
display('df_info',get_missing_columns(df_info))
display('df_item',get_missing_columns(df_item))
display('df_occupation',get_missing_columns(df_occupation))
display('df_test',get_missing_columns(df_test))
display('df_train',get_missing_columns(df_train))
display('df_user',get_missing_columns(df_user))
```
