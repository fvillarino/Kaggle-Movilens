```python
def get_mean_by_column(df,column, sales_str):
    group_means_dict = {}
    group_mean_list = []
    for columna, group_df in df.groupby('movieId'):
        group_mean =  group_df['rating'].mean()
        group_means_dict['rating'] = group_mean
        group_mean_list.append(group_mean)
    print('Return:', get_metric(df['rating'],df[column].apply(group_means_dict.get)))
    return group_means_dict, group_mean_list
```

```python
def get_metric(y, y_est):
    return np.sqrt((((y - y_est)/y)**2).sum()/len(y))
```

```python
from tensorflow.keras.utils import to_categorical
from matplotlib import pyplot as plt
from tensorflow.keras.callbacks import ModelCheckpoint 
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Input, Concatenate
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras import optimizers
from tensorflow.keras import initializers
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras import backend as K
```

```python
def rmspe(y_true, y_pred):
    return K.sqrt(K.mean(K.square((y_true - y_pred)/y_true)))
```

```python
hidden_units = 10
model = Sequential()
model.add(Dense(hidden_units, input_shape=(2,), name='Salida'))
model.add(Dense(1, activation='linear'))
model.summary()
```

```python
epochs = 100
model.compile(
    Adam(lr=0.1),
    loss='mse',
    metrics=[
        RootMeanSquaredError('rmse')
    ])

model.fit(X_train,y_train, epochs=epochs, batch_size=128)
```

```python
def get_keras_LR(X_columns, hidden_units=1):
    inputs = []
    activation = 'linear'
    if hidden_units>1:
        activation = 'relu'
    for i, col in enumerate(X_columns):
        inp = Input(shape=(X_train[i].shape[1],), name=f"{col}_input")
        inputs.append(inp)
    if len(X_columns)>1:
        concat_out = Concatenate()(inputs)
        dense_out = Dense(hidden_units, name='Dense', activation=activation)(concat_out)
    else:
        dense_out = Dense(hidden_units, name='Dense', activation=activation)(inputs[0])
    if hidden_units>1:
        dense_out = Dense(1, name='Dense_out')(dense_out)
    model = Model(inputs, dense_out)
    model.compile(optimizers.Adam(lr=0.0001), loss='mse', metrics=[rmspe, 'mse'])
    return model
```

```python
X_columns = list(X_train.columns)
X_train = np.hsplit(X_train[X_columns].values, len(X_columns))

model = get_keras_LR(X_columns, hidden_units=20)
model.summary()
```

```python
model.fit(X_train,y_train,batch_size=512, epochs=100)
```
