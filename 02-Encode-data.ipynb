```python
import pandas as pd
import numpy as np
from pandas_summary import DataFrameSummary
from sklearn_pandas import DataFrameMapper
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
PATH='Data/'
```

```python
df_train = pd.read_feather('preprocessed_train.fth')
df_test = pd.read_feather('preprocessed_test.fth')
```

```python
cat_vars = ['gender', 'occupation']
```

```python
cat_maps = [(o, LabelEncoder()) for o in cat_vars]
```

```python
mapper_cat = DataFrameMapper(cat_maps)
_ = mapper_cat.fit(df_train)
```

```python
mapper_cat
```

```python
N=1
print(list(zip(df_train['occupation'].values[:N], mapper_cat.transform(df_train)[:,1][:N])))
print(list(zip(df_train['gender'].values[:N], mapper_cat.transform(df_train)[:,0][:N])))
```

```python
#Asigno la conversion de las variables categoricas
df_train[cat_vars] = mapper_cat.transform(df_train)
df_test[cat_vars] = mapper_cat.transform(df_test)
```
