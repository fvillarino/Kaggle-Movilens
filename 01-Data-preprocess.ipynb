```python
# Hay que eliminar las peliculas duplicadas y modificar train y test

df_item_rep = df_item[df_item.drop('movieId',axis=1).duplicated()].sort_values(by='movieTitle')
item_rep_list = df_item_rep['movieTitle']

df_train = pd.read_csv(PATH + 'train.csv', sep='\t')
df_test = pd.read_csv(PATH + 'test.csv', sep=',')
```

```python
for title in item_rep_list:
    item_rep_ids_list = df_item[df_item['movieTitle']==title]['movieId']
    first_id = item_rep_ids_list.iloc[0]
    cmodificados=0
    for item_id in item_rep_ids_list:
        if(item_id!=first_id):
            df_train.loc[df_train['movieId']==item_id, 'movieId'] = first_id
            df_test.loc[df_test['movieId']==item_id, 'movieId'] = first_id
            cmodificados = cmodificados + 1 
    print('Id duplicado: '+str(first_id)+', Modificados: ' + str(item_rep_ids_list))
```

#### Transformo las fechas

```python
def add_datepart(df):
    df['release date'] = pd.to_datetime(df['release date'])
    df["Year"] = df['release date'].dt.year
    df["Month"] = df['release date'].dt.month
    df["Week"] = df['release date'].dt.week
    df["Day"] = df['release date'].dt.day
```
